```python
def standardize(x):
    return (x - x.mean())/x.std()
```

```python
def drop_rows_with_missing_data(df):
    return df[~df.isnull().apply(any, axis=1)]
```

```python
(surveys_df
     .pipe(drop_rows_with_missing_data)
     .assign(double_wt = surveys_df.weight * 2))
```
